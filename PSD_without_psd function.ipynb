```python
import numpy as np
import h5py
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
import scipy.signal
from scipy import *
import copy
from IPython.display import Audio
filename = "L-L1_GWOSC_O2_4KHZ_R1-1164689408-4096.hdf5"
f = h5py.File(filename, 'r')
o=f['strain/Strain']#Loading the original data
signal=np.array(o)#converting it to array
```

```python
for i in f['meta']:
    print(i,"=",str(f['meta/'+i][()]))
```

```python
srate=len(signal)//4096 #Duration=4096
print(srate)
```

```python
pnts = len(signal)
timevec = np.arange(0,pnts)/srate


# draw the time-domain signals
plt.plot(timevec,signal)
plt.xlabel('Time (s)')
plt.show()
```

```python
## static power spectrum and pick a frequency range

# inspect the power spectrum
hz = np.linspace(0,srate/2,len(signal)//2+1)
powr = abs(np.fft.fft(signal/pnts))
plt.plot(hz,powr[:int(len(hz))])
plt.xlim([100,2000])
plt.ylim([0,np.max(powr)])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

# Mean-smooth in frequency domain

```python
shape=(64,262144)
new=signal.reshape(shape)#creating matrix 2048x8192
l=[]
for i in range(len(new)):
    l.append(np.abs(np.fft.rfft(new[i]))**2)#new list after f4
```

```python
k=l[0]
for i in range(1,len(new)):
    k=k+l[i]
k=k/64
```

```python
plt.xlabel(r"Frequency")
plt.ylabel(r"Psd")
plt.title('PSD log log')
plt.loglog(k)
plt.plot(k)
```

```python
fltsig=np.real(np.fft.ifft(np.sqrt(k)))
```

```python
frex,time,tf = scipy.signal.spectrogram(signal,window=('tukey',.25),fs=srate,noverlap=100)

plt.pcolormesh(time,frex,np.log(tf))
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()
```

```python
frex,time,tf = scipy.signal.spectrogram(fltsig,window=('tukey',.25),fs=srate,noverlap=100)

plt.pcolormesh(time,frex,np.log(tf))
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()
```

```python
plt.xlabel(r"Frequency")
plt.ylabel(r"Psd")
plt.title('PSD log log')
f = np.linspace(0, 2048, len(k))
f2=np.linspace(0, 2048, len(np.abs(np.fft.rfft(signal))**2))
plt.loglog(f2,np.abs(np.fft.rfft(signal))**2)
plt.plot(f2,np.abs(np.fft.rfft(signal))**2,label='original')
plt.plot(f,k,label='Filtered')
plt.show()

```
